```python
def print_matrix(M):
    for r in range(len(M)):
        print("%2s" % str(r), end='')
        print("%2s" % ' rząd: ', end='')
        for c in range(len(M[0])):
            print("%5s" % str(M[r][c]), end='')
        print("\n")

import numpy as np

def knap_sack_01(weights, values, B):
    #B - pojemność plecaka
    #weights - tablica wag
    #values - tablica wartości
    
    n = len(values) #n - liczba przedmiotów
    K = [[None for _ in range(B + 1)] for _ in range(n + 1)] #macierz decyzji
    x = [0 for _ in range(n)] #tablica rozwiązania (1 - zabieramy przedmiot, 0 - nie zabieramy)
 
    for i in range(n + 1): #iteracja po przedmiotach
        for w in range(B + 1): #iteracja po wartościach sumarycznej wagi
            if i == 0 or w == 0:
                K[i][w] = 0 #przesunięcie indeksów
            elif weights[i - 1] <= w: #jeśli przedmiot się 'zmieści'
                K[i][w] = max(values[i - 1] + K[i - 1][w - weights[i - 1]], K[i - 1][w]) #decyzja co do zabrania przedmiotu bądź nie
            else:
                K[i][w] = K[i - 1][w]
    m = n
    Max = B
    while m != 0: #wypełnienie tablicy rozwiązań
        if K[m][Max] != K[m - 1][Max]:
            x[m - 1] = 1
            Max -= weights[m - 1]
        m -= 1

    weight_sum = 0 #wypełnienie sumarycznej wagi
    for i in range(len(x)):
        if x[i] == 1:
            weight_sum += weights[i]

    return K, K[n][w], x, weight_sum
```
