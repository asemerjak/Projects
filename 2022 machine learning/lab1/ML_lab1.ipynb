```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.datasets import load_iris
# load the famous iris data
irisRaw = load_iris()
```

```python
# read iris.data into a pandas DataFrame (df), including column names
iris = pd.DataFrame(data= np.c_[irisRaw['data'], irisRaw['target']],
                     columns= irisRaw['feature_names'] + ['target'])
```

```python
iris
```

```python
print("liczba wierszy = ", iris.shape[0])
print("liczba kolumn = ", iris.shape[1])
```

```python
iris.describe()
```

```python
iris.groupby("target").mean()
```

```python
iris.head()
```

```python
iris.dropna(inplace = True)
iris.sort_values(by=['sepal width (cm)'], inplace= True, ignore_index = True)
iris
```

```python
print("Minimalna długość płatka = ", iris.min(axis=0)[2])
print("Indeks minimalnej długości płatka = ", iris.idxmin(axis=0)[2])
print("Maksymalna długość płatka = ", iris.max(axis=0)[2])
print("Indeks maksymalnej długości płatka = ", iris.idxmax(axis=0)[2])
```

```python
iris.std()
```

```python
m = iris.mean()[0]
longer_irises = iris.loc[iris['sepal length (cm)'] > m].reset_index()
longer_irises
```
